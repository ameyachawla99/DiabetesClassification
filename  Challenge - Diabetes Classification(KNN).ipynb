```python
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
```

```python
def knn(x,y,query):
    vals=[]
    m=x.shape[0]
    for i in range(m):
        d=dist(query,x[i])
        vals.append((d,y[i]))
        
    vals=sorted(vals)
    vals=vals[:15]
    vals=np.array(vals)
    vals=np.unique(vals[:,1],return_counts=True)
    i=vals[1].argmax()
    return vals[0][i]
        
        
    
```

```python
a=knn(x,y,[6,160,88,42,321,38,0.786,39])
a
```

```python
m=test.shape[0]
Y=[]
for i in range(m):
    j=knn(x,y,test[i])
    Y.append(j)
```
